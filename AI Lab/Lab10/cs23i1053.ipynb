```python
# Note :
# I was getting a "wheel build error" and Microsoft C++ dependency issues
# while trying to install the older pomegranate library (0.14.9) on my system.
# To ensure the code runs without any errors, I have implemented the exact same
# Bayesian Network logic from scratch using pure Python. It directly calculates
# the Joint Probability Distribution P(a, m, i, e, s) using the chain rule and
# gives the exact same output as the library would.

import itertools

# ---------------------------------------------------------
# 1. ROOT NODES
# ---------------------------------------------------------
P_e = {'e0': 0.7, 'e1': 0.3}
P_i = {'i0': 0.8, 'i1': 0.2}

# ---------------------------------------------------------
# 2. CHILD NODES (Conditional Probabilities)
# ---------------------------------------------------------
# Marks (m) depends on IQ (i) and Exam (e)
P_m = {
    ('i0', 'e0', 'm0'): 0.6, ('i0', 'e0', 'm1'): 0.4,
    ('i0', 'e1', 'm0'): 0.9, ('i0', 'e1', 'm1'): 0.1,
    ('i1', 'e0', 'm0'): 0.5, ('i1', 'e0', 'm1'): 0.5,
    ('i1', 'e1', 'm0'): 0.8, ('i1', 'e1', 'm1'): 0.2
}

# Apti Score (s) depends on IQ (i)
P_s = {
    ('i0', 's0'): 0.75, ('i0', 's1'): 0.25,
    ('i1', 's0'): 0.40, ('i1', 's1'): 0.60
}

# Admission (a) depends on Marks (m)
P_a = {
    ('m0', 'a0'): 0.6, ('m0', 'a1'): 0.4,
    ('m1', 'a0'): 0.9, ('m1', 'a1'): 0.1
}

# ---------------------------------------------------------
# 3. CALCULATING JOINT PROBABILITY P(a, m, i, e, s)
# ---------------------------------------------------------

print("Calculating the Joint Probability Distribution P(a, m, i, e, s):\n")

variable_states = [['a0', 'a1'],
                   ['m0', 'm1'],
                   ['i0', 'i1'],
                   ['e0', 'e1'],
                   ['s0', 's1']]

combinations = list(itertools.product(*variable_states))

for combo in combinations:
    a, m, i, e, s = combo

    prob_e = P_e[e]
    prob_i = P_i[i]
    prob_m = P_m[(i, e, m)]
    prob_s = P_s[(i, s)]
    prob_a = P_a[(m, a)]

    # Bayesian Chain Rule calculation
    joint_prob = prob_a * prob_m * prob_i * prob_e * prob_s

    print(f"P({a}, {m}, {i}, {e}, {s}) = {joint_prob:.5f}")
```
